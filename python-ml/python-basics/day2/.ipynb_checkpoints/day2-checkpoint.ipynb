### jupyter lab commands
- esc: to go to command mode(block turns blue)
- enter: to go inside a block
- down arrow key: to go down inside a block
- esc+down: to go down in blocks
- ctrl+enter: run a block
- shift+enter: run and go down a block
- esc+m: convert a block to markdown block
- esc+b: to make a new block
- ctrl+a: select all
- ctrl+x: cut
- ctrl+c: copy
- ctrl+z: to undo a change
